## Splitting the dataframe into training and testing data

```python
#assigning independent variables to x and dependent variable or response variable to y
x = df.drop('class',axis = 1)
y = df['class']
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y, test_size = 0.20 , random_state = 100)
```

```python
print("Dimension of x_train :",x_train.shape)
print("Dimension of x_test :",x_test.shape)
print("Dimension of y_train :",y_train.shape)
print("Dimension of y_test :",y_test.shape)
```

## Building a Decision Tree model

```python
dt = DecisionTreeClassifier(random_state = 40)
dt.fit(x_train, y_train)
```

```python
dt_score_train = dt.score(x_train, y_train)
print("Training score: ",dt_score_train)
```

```python
dt_score_test = dt.score(x_test, y_test)
print("Testing score: ",dt_score_test)
```

```python
pred = dt.predict(x_test)
```

```python
print(confusion_matrix(y_test,pred))
```

#### It can be observed that accuracy score is 1 which clearly shows overfitting. In order to make a more generalised model k fold cross validation will be implemented

## K Fold Cross Validation

```python
#10-fold cross-validation
kfold = model_selection.KFold(n_splits = 5,random_state = 10,shuffle=True)
```

```python
result = model_selection.cross_val_score(dt,x_train,y_train,cv = kfold)
```

```python
print(result)
```

```python
print("Avg accuracy: {}".format(result.mean()))
```

#### Average accuracy score after implementing cross validation is 0.99
